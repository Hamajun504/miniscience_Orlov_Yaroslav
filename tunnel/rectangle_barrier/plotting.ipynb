```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 6.7082
l = 2.23607
a1, b1 = 0.39017 + 1.20057j, 0.75572 + 0.14994j
a2, b2 = -1.00299 + 0.12694j, 2.1489 + 1.2236j
x1 = np.arange(0, 1, 0.001)
x2 = np.arange(1, 1.2, 0.001)
x3 = np.arange(1.2, 2.2, 0.001)
data = np.genfromtxt("w02/U50.000000E45.000000000000.raw", delimiter=' ', skip_header=1).T
x = np.linspace(0, 2.2, 100000) 
plt.figure(figsize=(10,8))
plt.plot(x, np.sqrt(data[0]**2 + data[1]**2)[:-1], color="red", label="computation")
#plt.plot(x, data[0][:-1])
#plt.plot(x, data[1][:-1])
plt.plot(x1, abs(a1*np.exp(1j*k*(x1-1))+b1*np.exp(-1j*k*(x1-1))), color="blue", label="theory")
plt.plot(x2, abs(a2*np.exp(l*(x2-1))+b2*np.exp(-l*(x2-1))), color="blue")
plt.plot(x3, abs(np.exp(1j*k*(x3))), color="blue")
plt.legend()
plt.axvspan(1, 1.2, alpha=0.5, color='orange')
plt.savefig("w02/U50E45_abs.png",dpi=300)
```
