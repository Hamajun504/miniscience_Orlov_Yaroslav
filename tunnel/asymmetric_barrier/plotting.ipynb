```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_right = np.genfromtxt("results/rectangle_right000000.raw", delimiter=' ', skip_header=1).T
data_left = np.genfromtxt("results/rectangle_left000000.raw", delimiter=' ', skip_header=1).T
x = np.linspace(-1.1, 1.1, 100000) 
plt.figure(figsize=(10,8))
#plt.plot(x, np.sqrt(data[0]**2 + data[1]**2)[:-1])
plt.plot(x, data_right[0][:-1], label="right_re")
plt.plot(x, data_right[1][1:], label="right_im")

plt.plot(x, data_left[0][:-1], label="left_re")
plt.plot(x, data_left[1][:-1], label="left_im")
plt.legend()
plt.tight_layout()
plt.savefig("rectangle.png", dpi=300)
```

```python
data_right = np.genfromtxt("results/triangle_right000000.raw", delimiter=' ', skip_header=1).T
data_left = np.genfromtxt("results/triangle_left000000.raw", delimiter=' ', skip_header=1).T
x = np.linspace(-1.1, 1.1, 100000) 
plt.figure(figsize=(10,8))
plt.plot(x, (data_right[0]**2 + data_right[1]**2)[1:], label="right")
#plt.plot(x, data_right[0][:-1], label="right_re")
#plt.plot(x, data_right[1][:-1], label="right_im")

plt.plot(x, (data_left[0]**2 + data_left[1]**2)[:-1], label="left")
#plt.plot(x, data_left[0][:-1], label="left_re")
#plt.plot(x, data_left[1][:-1], label="left_im")
plt.legend()
plt.tight_layout()
plt.savefig("triangle.png", dpi=300)
```
